Ejercicio 1
Desarrolle un programa en Python que permita calcular el valor final del plan y el valor final de la radiografía a una clínica que ofrece descuentos según la edad y el quintil. 
Valores base a considerar: 
• Plan dental: $80.000 
• Radiografía dental: $12.000 
La siguiente tabla indica el descuento aplicado al plan: 



Además, la radiografía tiene reglas adicionales: 
• Quintil 1, 2 o 3: 10% 
• Si además edad ≥ 40: 5% adicional 
Si la edad es mayor a 45 años, el descuento del plan es 0%. 

Ejemplo 1: Ingrese edad: 42 
Ingrese quintil (1 a 5): 3 
El valor del plan es: 73600 
El valor de la radiografía es: 10200 

Ejemplo 2: Ingrese edad: 50 
Ingrese quintil (1 a 5): 4 
El valor del plan es: 80000 
El valor de la radiografía es: 12000 


```python
plan = 80000
radiografia_d = 12000

#Definir las variables y evitar errores
while True:
    print(" --- Bienvenido a CliniDuocUC --- ")
    print("Ingrese los siguientes datos para calcular su plan y radiografía.")
    try: 
        edad = int(input("Ingrese su edad: "))     
    except:
        print("Error. Debes ingresar un número entero.")
        continue
    break

while True:
    try: 
        quintil = int(input("Ingrese quintil (1-5): "))  

        if quintil < 1 or quintil > 5:
            print("Quintil fuera de rango. Intente otra vez.")
            continue
    except:
        print("Error. Debes ingresar un número entero.")
        continue
    break

#Ahora tengo mis 2 variables "edad" y "quintil" listas.

#Descuentos de plan
if edad > 45:
    print("No hay descuentos de plan")
elif edad <= 25 and (quintil == 1 or quintil == 2):
    plan = plan - (plan * 0.18)
elif edad <= 25 and (quintil == 3 or quintil == 4):
    plan = plan - (plan * 0.12)
elif (edad >= 26 and edad <= 45) and (quintil == 1 or quintil == 2):
    plan = plan - (plan * 0.12)
elif (edad >= 26 and edad <= 45) and (quintil == 3 or quintil == 4):
    plan = plan - (plan * 0.08)

#Descuentos de la radiografía
if (quintil == 1 or quintil == 2 or quintil == 3) and edad >= 40:
    radiografia_d = radiografia_d - (radiografia_d * 0.15)
elif quintil == 1 or quintil == 2 or quintil == 3:
    radiografia_d = radiografia_d - (radiografia_d * 0.10)


print(f"El valor del plan es: ${plan}")
print(f"El valor de la radiografía es: {radiografia_d}")
```

Ejercicio 2:

Desarrolle un programa en python que permita ingresar dos números enteros que indique un rango numérico. 
El primer valor debe ser menor que el segundo. 
Luego, el programa debe generar un número aleatorio (al azar) entre ese rango numérico. Para hacer esto, se sugiere lo siguiente: 
1. from random import randint 
2. numero = randint(num1, num2) 
La línea uno le permite cargar la función randint(). 
Luego, la línea dos usa la función randint() que permite generar un número aleatorio entre los valores que guardan las variables num1 y num2. 
A modo de ejemplo, randint(1,10), genera un número aleatorio entre 1 y 10. 

Una vez generado el número aleatorio, si el número resultara ser impar, el programa debe ajustarlo para que el número final a adivinar sea par, aplicando estas reglas: 
• Sí número + 1 no se sale del rango, entonces el número final será número + 1.
• Sí número + 1 se sale del rango, entonces el número final será número - 1. 

Con esto realizado, su programa debe simular un juego que permita adivinar el número final (ya ajustado). 
El juego consta de 3 intentos: 
• Si no se adivina en el primer intento, el programa debe indicar si el número a adivinar es mayor o menor. 
• Si aún no se adivina en el segundo intento, se debe indicar lo mismo y además dar una pista. La pista consiste en decir cuál intento estuvo más cerca del número correcto: si el número del primer intento o el del segundo intento.
• Si no se adivina en el tercer intento, se debe mostrar el mensaje: "Perdiste." y revelar el número. 
En caso de que se adivine el número (en cualquier intento), el programa debe terminar inmediatamente y mostrar: "Felicitaciones, pudiste adivinar.". 
Guíese por los ejemplos: 


```python
from random import randint

#Identificación de errores número 1
while True:
    try:
        num1 = int(input("Ingrese número 1: "))
    except:
        print("Error, debes ingresar un número entero.")
        continue
    break

#Identificación de errores número 2
while True:
    try:
        num2 = int(input("Ingrese número 2: "))
        #if para que el num2 no sea menor al num1
        if num2 <= num1:
            print("El segundo número no puede ser menor o igual al primero.")
            continue
    except:
        print("Error, debes ingresar un número entero.")
        continue
    break

#Variable de número alteatorio entre num1 y num2
numero = randint(num1, num2)
print(numero)

# num%2 me entrega el residuo de una división. 
# Si me da 0 es que el número es par. Ej: 10%2: 0 --> porque no sobraría nada
# Si me da 1 el número sería impar. Ej: 9%2: 1 --> sobra 1. Porque lo máximo en que se puede dividir es 8 (2x4 = 8) y nos sobra 1
if numero%2 == 0:
    print("Número par")
else:
    print("Número impar")
    print(numero)

    #Si el número está dentro del rango (menor al num2) se le sumará 1
    if (numero+1) <= num2:
        numero += 1
    #Si el número está fuera del rango (mayor al num2) se le restará 1
    elif (numero+1) > num2:
        numero -= 1

    print(numero)

while True:
    #Primer intento
    intento1 = int(input("Intente adivinar (1er intento): "))
    if intento1 == numero:
        print(f"Felicitaciones, pudiste adivinar! El número secreto es {numero}")
        break
    else:
        if numero < intento1:
            print("El número es menor.")
        else:
            print("El número es mayor.")

    #Segundo intento
    intento2 = int(input("Intente adivinar (2do intento): "))

    if intento2 == numero:
        print(f"Felicitaciones, pudiste adivinar! El número secreto es {numero}")
        break
    else:
        if numero < intento2:
            print("El número es menor")
        else:
            print("El número es mayor.")
    

    #Para la pista necesitamos calcular las distancias que hay entre los números primero

    #Calcular distancia intento1
    if intento1 > numero:
        distancia1 = intento1 - numero
    else:
        distancia1 = numero - intento1

    #Calcular distancia intento2
    if intento2 > numero:
        distancia2 = intento2 - numero
    else:
        distancia2 = numero - intento2

    #Ver cual es las 2 distancias está más cerca
    if distancia1 < distancia2:
        print(f"El número que buscas está más cerca de {intento1} que de {intento2} ")
    else:
        print(f"El número que buscas está más cerca de {intento2} que de {intento1}")

    
    #Tercer intento
    intento3 = int(input("Intente adivinar (último intento): "))
    if intento3 != numero:
        print("Perdiste")
        break
    else:
        print(f"Felicitaciones, pudiste adivinar! El número secreto es {numero}")
        break
    

    




```
